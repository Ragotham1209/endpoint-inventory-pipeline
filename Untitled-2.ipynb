# Global Application Inventory Pipeline (Polars + Parquet)

A data pipeline that:
1. Reads SCCM/DesktopOne **Parquet** source files
2. Standardises application names via declarative mapping rules
3. Cross-joins machines × apps to produce a compliance matrix
4. Writes the result as **Parquet** (compressed, typed, fast)

```python
# ── Imports, Configuration & Mapping Rules ──────────────────────────
from __future__ import annotations

import logging
from pathlib import Path
from typing import NamedTuple

import polars as pl

# ── Logging ──────────────────────────────────────────────────────────
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s",
)
log = logging.getLogger(__name__)

# ── Paths ────────────────────────────────────────────────────────────
FOLDER_PATH = Path(r"\\servername\foldername")
OUTPUT_FILE = FOLDER_PATH / "Global_Application_Inventory_Master.parquet"

# ── Parquet file glob patterns ────────────────────────────────────────
FILE_PATTERNS: dict[str, str] = {
    "sys_info":  "DesktopOne_Sys_Info_0-F*.parquet",
    "teams":     "DesktopOne_Teams_0-F*.parquet",
    "onedrive":  "DesktopOne_OneDrive_0-F*.parquet",
    "partition": "DesktopOne_Partition_*.parquet",
}


# ── Declarative mapping rules ────────────────────────────────────────
class MappingRule(NamedTuple):
    """Single product-name mapping rule."""
    pattern: str          # regex applied to ProductName
    exact: bool           # True → equality check instead of contains
    mapped_name: str
    app_key: str


MAPPING_RULES: list[MappingRule] = [
    # ─── Core applications ───
    MappingRule(r"(?i)Adobe Acrobat",                                 False, "Adobe Acrobat",                                "Adobe Acrobat Patch"),
    MappingRule(r"(?i).*LogMeIn Rescue (Calling Card|Remote Assistance).*", False, "LogMeIn Rescue",                         "LogMeIn Rescue"),
    MappingRule(r"(?i)Microsoft Edge",                                False, "Microsoft Edge",                               "Microsoft Edge"),
    MappingRule(r"(?i)Visual C\+\+.*v14.*x86",                        False, "Microsoft Visual C++ v14 Redistributable x86", "Microsoft Visual C++ v14 Redistributable x86"),
    MappingRule(r"(?i)Visual C\+\+.*v14.*x64",                        False, "Microsoft Visual C++ v14 Redistributable x64", "Microsoft Visual C++ v14 Redistributable x64"),
    MappingRule(r"(?i)Steelray Project Viewer",                       False, "Steelray Project Viewer",                      "Steelray Project Viewer"),
    MappingRule(r"(?i)WinZip (Command Line|C)",                       False, "WinZip Command Line",                          "WinZip Command Line"),
    MappingRule(r"(?i)WinZip.*[0-9]",                                 False, "WinZip Enterprise",                            "WinZip Enterprise"),
    MappingRule(r"(?i)Microsoft Teams",                               False, "Microsoft Teams (New)",                         "Microsoft Teams (New)"),
    MappingRule(r"(?i)Microsoft OneDrive",                            False, "Microsoft OneDrive for Business",               "Microsoft OneDrive for Business"),

    # ─── Optional applications ───
    MappingRule(r"(?i)^Druva inSync.*",                               False, "Druva inSync",                                 "Druva InSync"),
    MappingRule("BIG-IP Edge Client",                                  True, "BIG-IP Edge Client",                           "F5Networks BIGIP Client"),
    MappingRule("Google Chrome",                                       True, "Google Chrome",                                 "Google Chrome x64"),
    MappingRule(r"(?i)^Microsoft PowerBi Desktop.*",                  False, "Microsoft PowerBi Desktop",                    "Microsoft PowerBi Desktop"),
    # SQL Server
    MappingRule(r"(?i)^Microsoft SQL Server 2017( \(64-bit\)| Setup \(English\))?$", False, "Microsoft SQL Server 2017",     "Microsoft SQL Server 2017"),
    MappingRule(r"(?i)Microsoft SQL Server 2019(?!.*CTP)",            False, "Microsoft SQL Server 2019",                    "Microsoft SQL Server 2019"),
    MappingRule(r"(?i)Microsoft.*SQL.*Server 2022",                   False, "Microsoft SQL Server 2022",                    "Microsoft SQL Server 2022"),
    # Firefox
    MappingRule(r"(?i)^Mozilla Firefox.*",                            False, "Mozilla Firefox",                              "Mozilla Firefox x64"),
    # Java / JDK
    MappingRule(r"(?i)Java.*Update.*\([0-9]+\)?.*64",                 False, "Oracle Java Runtime x64",                      "Oracle Java Runtime x64"),
    MappingRule(r"(?i)Java.*[0-9].*Update.*[0-9]+(?!.*64)",           False, "Oracle Java Runtime x86",                      "Oracle Java Runtime x86"),
    MappingRule(r"(?i)Java.*SE Development Kit.*64",                  False, "Oracle JDK x64",                               "Oracle JDK x64"),
    MappingRule(r"(?i)Java.*SE Development Kit(?!.*64)",              False, "Oracle JDK x86",                               "Oracle JDK x86"),
    # Security & Collaboration
    MappingRule("GlobalProtect",                                       True, "Palo Alto Networks Global Protect x64",        "Palo Alto Networks Global Protect x64"),
    MappingRule(r"(?i)^Zoom( VDI)? Workplace.*64|^Zoom \(64-bit\)$|^Zoom\(64bit\)$", False, "Zoom Workplace x64",           "Zoom Workplace x64"),
    MappingRule("Zoom Outlook Plugin",                                 True, "Zoom Outlook Plugin",                          "Zoom Outlook for Plugin"),
]

log.info("Configuration loaded — %d mapping rules defined.", len(MAPPING_RULES))
```

```python
# ── Helper Functions ─────────────────────────────────────────────────

def load_csv(pattern: str, *, folder: Path = FOLDER_PATH) -> pl.LazyFrame:
    """
    Read Windows-generated CSV exports (UTF-16 with BOM), falling back to
    UTF-8.  Headers are sanitised to remove null bytes and stray quotes.
    """
    source = folder / pattern
    for encoding in ("utf16", "utf8-lossy"):
        try:
            lf = pl.scan_csv(
                source,
                encoding=encoding,
                infer_schema_length=10_000,
                truncate_ragged_lines=True,
            )
            # Strip \0 and " artefacts from column names
            raw_cols = lf.collect_schema().names()
            rename_map = {
                c: c.replace("\0", "").replace('"', "").strip()
                for c in raw_cols
            }
            log.info("Loaded '%s' (%s) — %d columns.", pattern, encoding, len(raw_cols))
            return lf.rename(rename_map)
        except Exception:
            continue

    raise RuntimeError(f"Could not read {source} with any supported encoding.")


def build_mapping_expr(rules: list[MappingRule]) -> pl.Expr:
    """
    Convert the flat list of MappingRules into a chained
    when/then/otherwise Polars expression returning a struct
    {MappedName, AppKey} or null.
    """
    expr = pl.when(pl.lit(False)).then(None)  # seed — never matches

    for rule in rules:
        condition = (
            pl.col("ProductName") == rule.pattern
            if rule.exact
            else pl.col("ProductName").str.contains(rule.pattern)
        )
        expr = expr.when(condition).then(
            pl.struct(
                MappedName=pl.lit(rule.mapped_name),
                AppKey=pl.lit(rule.app_key),
            )
        )

    return expr.otherwise(None).alias("MappedData")


log.info("Helper functions ready.")
```

```python
# ── Pipeline: Extract → Map → Gap-fill → Parquet ────────────────────

# ── 1. Master dimension (unique machines) ───────────────────────────
log.info("Step 1 — Loading master machine dimension")

sys_info = (
    load_csv(FILE_PATTERNS["sys_info"])
    .select("NetBiosName", "ResourceID")
    .unique(subset=["NetBiosName"])
)

# ── 2. Load & standardise inventory streams ─────────────────────────
log.info("Step 2 — Loading inventory streams")

COMMON_COLS = ["NetBiosName", "ResourceID", "ProductName", "Version"]

df_partition = load_csv(FILE_PATTERNS["partition"]).select(COMMON_COLS)
df_teams     = load_csv(FILE_PATTERNS["teams"]).select(COMMON_COLS)
df_onedrive  = (
    load_csv(FILE_PATTERNS["onedrive"])
    .select(
        "NetBiosName",
        "ResourceID",
        "ProductName",
        pl.col("Version0").alias("Version"),
    )
)

raw_inventory = pl.concat([df_partition, df_teams, df_onedrive], how="vertical")

# ── 3. Apply global mapping rules ───────────────────────────────────
log.info("Step 3 — Applying %d mapping rules", len(MAPPING_RULES))

mapping_expr = build_mapping_expr(MAPPING_RULES)

global_mapped = (
    raw_inventory
    .with_columns(mapping_expr)
    .filter(pl.col("MappedData").is_not_null())
    .drop("ProductName")
    .unnest("MappedData")
    .rename({"MappedName": "ProductName"})
    .unique(subset=["NetBiosName", "AppKey"])
)

# ── 4. Gap analysis — every machine × every app ─────────────────────
log.info("Step 4 — Generating compliance matrix (cross join)")

all_app_keys = global_mapped.select("AppKey").unique().collect().lazy()

expected_grid = sys_info.join(all_app_keys, how="cross")

df_final_lazy = (
    expected_grid
    .join(global_mapped, on=["NetBiosName", "AppKey"], how="left")
    .with_columns(
        pl.col("ProductName").fill_null(pl.col("AppKey")),
        pl.col("Version").fill_null(pl.lit("Not Installed")),
    )
    .select("NetBiosName", "ResourceID", "AppKey", "ProductName", "Version")
)

# ── 5. Collect, audit & write Parquet ────────────────────────────────
log.info("Step 5 — Collecting results & writing Parquet")

try:
    df_final = df_final_lazy.collect()
    total_machines = sys_info.select(pl.len()).collect().item()

    log.info(
        "Pipeline complete — %d machines, %d records",
        total_machines,
        len(df_final),
    )

    # Per-app compliance audit
    audit = (
        df_final
        .with_columns(
            pl.when(pl.col("Version") == "Not Installed")
            .then(pl.lit("Missing"))
            .otherwise(pl.lit("Installed"))
            .alias("Status")
        )
        .group_by("AppKey")
        .agg(
            (pl.col("Status") == "Installed").sum().alias("Installed"),
            (pl.col("Status") == "Missing").sum().alias("Missing"),
            pl.len().alias("Total_Checked"),
        )
        .sort("AppKey")
    )

    print("\nAPP COMPLIANCE AUDIT")
    print("-" * 65)
    with pl.Config(tbl_rows=100):
        print(audit)
    print("-" * 65)

    # Write Parquet (zstd compression for best size/speed trade-off)
    df_final.write_parquet(OUTPUT_FILE, compression="zstd")
    log.info("Parquet saved → %s", OUTPUT_FILE)

except Exception:
    log.exception("Pipeline failed")
```
